```python
A = torch.from_numpy(np.array([[2., 3.], [1., 0]]))
A.requires_grad_(True)
B = torch.from_numpy(np.array([[1., 2.], [3.,4.]]))
B.requires_grad_(True)

C = torch.matmul(A,B)
```

```python
d = torch.from_numpy(np.array([[3.][2.]]))
d.requires_grad_(True)
```

```python
E = C + d

J = torch.sum(E)
```
